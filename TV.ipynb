```python
# MBC 방영 프로그램

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3)

title = []
likes = []
genre = []

for i in range(0,3) :
    entext = ['drama', 'ent', 'culture']
    driver.get("http://www.imbc.com/broad/tv/"+ entext[i] +"/")
    time.sleep(0.3)

    if i == 2 :
        ttt = ['current','cultural','special']
        tttn = ['시사', '교양', '특집']
        for j in range(0,3,1) :
            taburl = '#container > div.con-area.list-area > div.list-bottom.wrapper > div.culture-tab > button.ct-tab.'+str(ttt[j])+'> span.txt'
            tab = driver.find_element_by_css_selector(taburl)
            tab.click()
            time.sleep(0.5)
            
            titleList = driver.find_elements_by_css_selector('#container > div.con-area.list-area > div.list-bottom.wrapper > ul > li > div.box-wrap > div.txt > div.tit')
            time.sleep(0.3)
            likeList = driver.find_elements_by_css_selector('#container > div.con-area.list-area > div.list-bottom.wrapper > ul > li > div.box-wrap > div.img > div > button > span:nth-child(2)')
            time.sleep(0.3)
            genreList = [entext[i]+'_'+tttn[j]] *len(titleList)
            genre.extend(genreList)
    
            for tit in titleList :
                title.append(tit.text)
            for ht in likeList :
                content = ht.text
                content = re.sub(",", '', content)
                content = content.strip()
                likes.append(content)
            time.sleep(0.3)
    
    else :    
        titleList = driver.find_elements_by_css_selector('#container > div.con-area.list-area > div.list-bottom.wrapper > ul > li > div.box-wrap > div.txt > div.tit')
        time.sleep(0.3)
        likeList = driver.find_elements_by_css_selector('#container > div.con-area.list-area > div.list-bottom.wrapper > ul > li > div.box-wrap > div.img > div > button > span:nth-child(2)')
        time.sleep(0.3)
        genreList = [entext[i]] *len(titleList)
        genre.extend(genreList)
        
        for tit in titleList :
            title.append(tit.text)
        for ht in likeList :
            content = ht.text
            content = re.sub(",", '', content)
            content = content.strip()
            likes.append(content)  
        time.sleep(0.3)


with open("output/MBC방영프로그램.csv","wt", encoding="utf-8") as f :
    f.write('genre,title,like\n')
    for i in range(len(genre)):
        f.write(genre[i]+","+title[i]+","+likes[i]+'\n')
```

```python
# 장르 검색

df = pd.read_csv("output/TV_통합.csv")
df['genre'] = 0

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3)
driver.get("https://www.naver.com")
time.sleep(0.3)


for i in range(0,len(df.index)) :
    if df.loc[i, 'title'].find('드라마') != -1 or df.loc[i, 'title'].find('미니시리즈') != -1 or df.loc[i, 'title'].find('일일극') != -1 :
        df.loc[i, 'genre'] = '드라마'
    elif df.loc[i,'title'].find('뉴스') != -1 :
        df.loc[i, 'genre'] = '뉴스'
    else :
        a= df.loc[i,'title']
        driver.get("https://www.naver.com")
        time.sleep(0.3)
        target = driver.find_element_by_css_selector("[name = 'query']")
        target.send_keys(str(a))
        target.send_keys(Keys.ENTER)
        time.sleep(0.3)
        try:
            txt = driver.find_element_by_css_selector("#main_pack > div.sc_new._kgs_broadcast.cs_common_module.case_normal > div.cm_top_wrap._sticky._custom_select > div.title_area._title_area > div.sub_title > span:nth-child(1)")
            df.loc[i,'genre'] = txt.text
            time.sleep(0.3)
        except :
            df.loc[i,'genre'] = '기타'
            time.sleep(0.3)
            
            
with open("output/TV통합장르.csv","wt", encoding="utf-8") as f :
    f.write('genre\n')
    for i in range(0,len(df.index)):
        f.write(df.loc[i,'genre']+'\n')
```

```python
# 예능 안에서의 분류
ndf = pd.read_csv("output/TV_통합_이후.csv")
ndf = ndf.loc[(ndf.genre == '예능'),[ 'number', 'detail', 'genre']]

dfgrouped = pd.DataFrame(ndf['number'].groupby(ndf['detail']).sum())
n = ndf['number'].sum()
dfgrouped['percent'] = round(dfgrouped['number']/n*100,1)

ratio = dfgrouped['percent']
labels = dfgrouped.index
picolors = sns.color_palette('tab20')

fig = plt.figure(figsize=(20,15))
fig.set_facecolor('white')
ax = fig.add_subplot()
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 3}
pie = ax.pie(ratio, startangle=90, counterclock=False, 
        colors=picolors, wedgeprops=wedgeprops)
title_font = {'fontsize': 30, 'fontweight': 'bold', 'color': 'purple'}
plt.title("2021년 예능 속 장르 선호도 비율", fontdict=title_font)
total = np.sum(ratio)

threshold = 0.85 ## 상한선 비율
sum_pct = 0 ## 퍼센티지
bbox_props = dict(boxstyle='square',fc='w',ec='w',alpha=0) ## annotation 박스 스타일
 
## annotation 설정
config = dict(arrowprops=dict(arrowstyle='-'),bbox=bbox_props,va='center')
 
for i,l in enumerate(labels):
    ang1, ang2 = ax.patches[i].theta1, ax.patches[i].theta2 ## 파이의 시작 각도와 끝 각도
    center, r = ax.patches[i].center, ax.patches[i].r ## 원의 중심 좌표와 반지름길이
    
    if i < len(labels) - 1:
        sum_pct += float(f'{ratio[i]/total*100:.2f}')
        text = f'{ratio[i]/total*100:.1f}%'
    else: ## 마지막 파이 조각은 퍼센티지의 합이 100이 되도록 비율을 조절
        text = f'{100-sum_pct:.2f}%'
    
    ## 비율 상한선보다 작은 것들은 Annotation으로 만든다.
    if ratio[i]/total*100 < threshold:
        ang = (ang1+ang2)/2 ## 중심각
        x = np.cos(np.deg2rad(ang)) ## Annotation의 끝점에 해당하는 x좌표
        y = np.sin(np.deg2rad(ang)) ## Annotation의 끝점에 해당하는 y좌표
        
        horizontalalignment = {-1: "left", 1: "right"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang) ## 시작점과 끝점 연결 스타일
        config["arrowprops"].update({"connectionstyle": connectionstyle}) ## 
        ax.annotate(text, xy=(x, y), xytext=(1.5*x, 1.2*y),
                    horizontalalignment=horizontalalignment, **config)
    else:
        x = (r/2)*np.cos(np.pi/180*((ang1+ang2)/2)) + center[0] ## 텍스트 x좌표
        y = (r/2)*np.sin(np.pi/180*((ang1+ang2)/2)) + center[1] ## 텍스트 y좌표
        ax.text(x,y,text,ha='center',va='center',fontsize=16)
        
plt.legend(labels , bbox_to_anchor=(1.05, 1))
plt.savefig("project/예능장르선호도소분류비율.png") 
```
